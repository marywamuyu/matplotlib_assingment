```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
data=('month','revenue','order','avarege_order_value')
```

```python
df=pd.read_csv('jumta.csv')
```

```python
df.head
```

```python
pd.to_numeric
df.dropna()
```

# line plot

```python
plt.figure(figsize=(15,10))
plt.plot(df['Month'], df['Revenue'], marker='o', linestyle='--', linewidth=3, label='Rev')
plt.title('Monthly Revenue ')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

# bar plot

```python
plt.figure(figsize=(10,5))
plt.bar(df['Month'], df['Orders'], color='black', label='Orders')
plt.title('orders per month')
plt.xlabel('Month')
plt.ylabel('orders per month')
plt.grid()
plt.legend()
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(8,8))
plt.scatter(x='Orders', y='Average_Order_Value', color='red', label='Order Value', data=df)
plt.title('orders against average orders')
plt.ylabel('Avrg_order_val', fontsize=18)
plt.xlabel('Orders', fontsize=18)
plt.legend()
plt.show()

```

# subplot

```python
fig, ax = plt.subplots(3, 1, figsize=(10, 16))


ax[0].plot(df['Month'], df['Revenue'], marker='o', linestyle='-', color='blue')
ax[0].set_title('Monthly Revenue Over Two Years')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Revenue')
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(True)


ax[1].bar(df['Month'], df['Orders'], color='green')
ax[1].set_title('Total Number of Orders Per Month')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Number of Orders')
ax[1].tick_params(axis='x', rotation=45)
ax[1].grid(True)


ax[2].scatter(df['Orders'], df['Average_Order_Value'], color='red')
ax[2].set_title('Relationship between Orders and Average Order Value')
ax[2].set_xlabel('Number of Orders')
ax[2].set_ylabel('Average Order Value')
ax[2].grid(True)


plt.tight_layout()


plt.savefig('publication-quality-plot.png', dpi=300)
```
